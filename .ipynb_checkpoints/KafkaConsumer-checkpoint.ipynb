```python
import pandas as pd
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
import s3fs
import os
from dotenv import load_dotenv
import boto3
from datetime import datetime
load_dotenv()
```

```python
consumer = KafkaConsumer('demo_testing3',
                          bootstrap_servers=['18.224.151.151:9092'], group_id=None,    enable_auto_commit=False ,
                          value_deserializer=lambda x:
                          loads(x.decode('utf-8')))
```

```python
class s3_bucket:
    def __init__(self):
        self.s3resource = boto3.resource('s3')
        self.bucket_name = 'kafkabucketstock'
        self.bucket=self.s3resource.Bucket(self.bucket_name)
        
    def upload_s3_bucket(self, string_text,current_batch):
        self.current_batch=current_batch
        self.timestr = datetime.now().strftime("%Y%m%d-%H%M%S.%f")  
        self.bucket.put_object(Key=f'example_{self.timestr}_{self.current_batch}.json',Body=string_text)
        print(f'example_{self.timestr}_{self.current_batch}.json')
s3=s3_bucket()
```

```python
json_string='{}'
batch_size = 100
current_batch = 1
list_data=[]
file_name_counter=0
try:

    for i in consumer:
        if current_batch<batch_size:
            list_data.append(i.value)
            current_batch=current_batch+1
            file_name_counter=file_name_counter+1
        else:
            list_data.append(i.value)
            json_string='{}'
            json_string=json.dumps(list_data, indent=2)
            print(f"Uploading batch: {list_data}")
            s3.upload_s3_bucket(json_string,current_batch)
            consumer.commit()
            list_data=[]
            current_batch=1
    
    if list_data:
        json_string='{}'
        json_string=json.dumps(list_data, indent=2)
        print(f"Uploading batch: {list_data}")
        s3.upload_s3_bucket(json_string,current_batch)
        consumer.commit()
        
except Exception as e:
    print(f"Failed to upload batch: {e}")
```
