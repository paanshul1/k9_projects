```python
class s3_bucket:
    def __init__(self):
        self.s3resource = boto3.resource('s3')
        self.bucket_name = 'kafkabucketstock'
        self.bucket=self.s3resource.Bucket(self.bucket_name)
        
    def upload_s3_bucket(self, string_text,current_batch):
        self.current_batch=current_batch
        self.timestr = datetime.now().strftime("%Y%m%d-%H%M%S.%f")  
        self.bucket.put_object(Key=f'example_{self.timestr}_{self.current_batch}.json',Body=string_text)
        print(f'example_{self.timestr}_{self.current_batch}.json')
s3=s3_bucket()
```

```python
import json
import time
#from kafka import KafkaConsumer

 

json_string = '{}'
list_data = []
file_name_counter = 0
exist_flag=0
try:
    consumer.subscribe(['demo_testing3'])
    print('in first try block')
    while True:
        message=consumer.poll(timeout_ms=10000, max_records= 100)
        if not message:
            print('There is no message ... ')
            break;
        for tp,records in message.items():
            for i in records:
                list_data.append(i.value)
                if len(list_data)>=100:
                    json_string = json.dumps(list_data, indent=2)
                    print(f"Uploading batch: {list_data}")
                    file_name_counter += 1
                    print(f'file_name_counter: {file_name_counter}')
                    s3.upload_s3_bucket(json_string, file_name_counter)
                    list_data = []

except Exception as e:
    print(f"Failed to upload batch: {e}")
finally:
     if len(list_data)>0:
        print('in finally')
        json_string = json.dumps(list_data, indent=2)
        print(f"Uploading incomplete batch: {list_data}")
        file_name_counter += 1
        print(f'file_name_counter: {file_name_counter}')
        s3.upload_s3_bucket(json_string,'ZERO'+str(file_name_counter)+'inc')
        list_data = []
        

 
```
